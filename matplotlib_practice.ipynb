```python
# Import dependencies.
#The pyplot module will allow us to create all the necessary charts; create a plotting area in a
#figure; plot lines in a plotting area; and annotate the chart with labels, colors, and other decorative features.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
# plt.show() should only be used once per session because it can cause the graph to display improperly. If you have problems with data not being displayed properly,
#you may have to restart your kernel and clear your output.
plt.plot(x_axis, y_axis)
plt.show()
```

## When we switch from using MATLAB to the object-oriented interface method, we use ax.plot() instead of plt.plot()

### To create a figure, use fig, ax = plt.subplots().
To define the axes, use ax.plot().
Inside the parentheses of ax.plot(), add the axes separated by a comma.

```python
# Let's recreate the line chart we just made using the object-oriented interface method. In a new cell, add the following code:

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# he fig, ax = plt.subplots() line of code can also be written like this.
# Create the plot with ax.plt()
#To change figure-level attributes
#(such as axis labels, a title, or a legend) or save the figure as an image, use fig = plt.figure()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis) 
```

## Create a Horizontal Bar Chart

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

### To invert the y-axis to have the months in ascending order, use the gca() method. The gca() method means "get current axes." We can chain the gca() method to the invert_yaxis() method by using gca().invert_yaxis(), as shown here

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

## Create Scatter Plots and Bubble Charts

###  A scatter plot visually depicts a positive, negative, or no-correlation relationship between two factors.

### Create a Scatter Plot Using the MATLAB Method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

### When we use the plt.plot() function to create a scatter plot, we need to add a lowercase "o" as a parameter inside the parentheses. This switches the plot from a line chart to a scatter plot.

```python
plt.plot(x_axis, y_axis, 'o')
```

### To create a scatter plot using the plt.scatter() function, add the x-axis and y-axis parameters inside the parentheses, as shown here

```python
plt.scatter(y_axis, x_axis)
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(5, 50)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

##### We can refactor the code we used to create a scatter plot above, but instead of using a for loop to create the list y_axis_larger for the size parameter of the marker, we can use Python's list comprehension technique inside the plt.scatter() function. You can use list comprehension to replace many for and while loops. List comprehension is faster because it is optimized for Python to spot a predictable pattern during looping.

```python
#format
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

## Create a Bubble Chart Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
#ax.scatter(x_axis, y_axis, s=y_axis, color="sky blue", label="Boston")
plt.scatter(y_axis, x_axis, color="xkcd:sky blue", label='Boston', s=y_axis)
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(5, 50)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.gca().invert_yaxis()
```

## Create Pie Charts

### Create Pie Charts Using the MATLAB Method

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, .1, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
